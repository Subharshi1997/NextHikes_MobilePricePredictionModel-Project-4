```python
# ============================================================
# SMARTPHONE PRICE PREDICTION — FULL ML PIPELINE v4
# ============================================================
#
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import (
    train_test_split, RandomizedSearchCV, cross_val_score
)
from sklearn.preprocessing import RobustScaler, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.feature_selection import SelectKBest, mutual_info_regression
import xgboost as xgb
import ipywidgets as widgets
from IPython.display import display, clear_output, Markdown
from statsmodels.stats.outliers_influence import variance_inflation_factor
from scipy.stats import randint, uniform
import warnings
warnings.filterwarnings('ignore')

sns.set_theme(style='whitegrid', palette='muted')

# ═══════════════════════════════════════════════════════════
# STAGE 1 — DATA LOADING & INITIAL CLEANING
# ═══════════════════════════════════════════════════════════
print("=" * 60)
print("STAGE 1: DATA LOADING & INITIAL CLEANING")
print("=" * 60)

df = pd.read_excel('Processed_Flipdata (1).xlsx')

# Rename FIRST — before any exploration or display
df.columns = df.columns.str.strip()
df.rename(columns={
    'Battery_': 'Battery',
    'Processor_': 'Processor',
    'Prize': 'Price'
}, inplace=True)

if 'Unnamed: 0' in df.columns:
    df.drop('Unnamed: 0', axis=1, inplace=True)

print(f"Shape: {df.shape}")
display(df.head())
df.info()
display(df.describe(include='all'))
print("\nMissing Values:")
print(df.isnull().sum())

# ═══════════════════════════════════════════════════════════
# STAGE 2 — FEATURE ENGINEERING
# ═══════════════════════════════════════════════════════════
print("\n" + "=" * 60)
print("STAGE 2: FEATURE ENGINEERING")
print("=" * 60)

#  Brand from Model name
df['Brand'] = df['Model'].apply(
    lambda x: x.split()[0].upper() if isinstance(x, str) else 'UNKNOWN'
)
df['Brand'] = df['Brand'].replace({
    'MICROMAX1': 'MICROMAX',
    'REDMI': 'XIAOMI',
    'redmi': 'XIAOMI'
})

# Camera megapixels — robust to "50MP" and "50 MP" formats
def extract_mp(cam):
    if isinstance(cam, str):
        cleaned = cam.upper().replace(' ', '')
        if 'MP' in cleaned:
            try:
                return int(cleaned.replace('MP', '').strip())
            except ValueError:
                return 0
    return 0

df['Rear_MP']  = df['Rear Camera'].apply(extract_mp)
df['Front_MP'] = df['Front Camera'].apply(extract_mp)

# Processor brand
def extract_processor_brand(proc):
    p = str(proc).lower()
    if 'qualcomm'   in p or 'snapdragon' in p:                      return 'Qualcomm'
    elif 'mediatek' in p or 'dimensity'  in p or 'helio' in p \
         or 'mtk'   in p:                                            return 'Mediatek'
    elif 'unisoc'   in p or 'spreadtrum' in p:                       return 'Unisoc'
    elif 'exynos'   in p or 'sec'        in p:                       return 'Exynos'
    elif 'google'   in p or 'tensor'     in p:                       return 'Google'
    elif 'apple'    in p or 'ios'        in p:                       return 'Apple'
    else:                                                             return 'Other'

df['Processor_Brand'] = df['Processor'].apply(extract_processor_brand)

# Processor tier (string label + numeric for interactions)
def get_processor_tier(proc):
    p = str(proc).lower()
    premium_kw = ['snapdragon 8', 'dimensity 9', 'dimensity 13', 'tensor',
                  'exynos 2', 'snapdragon 888', 'snapdragon 8 gen']
    mid_kw     = ['snapdragon 7', 'dimensity 8', 'dimensity 7', 'helio g9',
                  'helio g8', 'snapdragon 6', 'exynos 1', 'helio g']
    if any(k in p for k in premium_kw): return 'Premium'
    elif any(k in p for k in mid_kw):   return 'Mid'
    else:                                return 'Budget'

df['Processor_Tier']     = df['Processor'].apply(get_processor_tier)
tier_map                 = {'Budget': 1, 'Mid': 2, 'Premium': 3}
df['Processor_Tier_Num'] = df['Processor_Tier'].map(tier_map)

# 5G flag
df['Is_5G'] = (
    df['Model'].str.contains('5G', na=False) |
    df['Processor'].str.contains('5G', na=False)
).astype(int)

# AI Lens binary flag (if column exists)
if 'AI Lens' in df.columns:
    df['Has_AI_Lens'] = df['AI Lens'].notna().astype(int)
    df.drop('AI Lens', axis=1, inplace=True)
    print("  Engineered Has_AI_Lens from 'AI Lens' column.")

print("Engineered: Brand, Rear_MP, Front_MP, Processor_Brand,")
print("  Processor_Tier, Processor_Tier_Num, Is_5G")

# ═══════════════════════════════════════════════════════════
# STAGE 3 — OUTLIER REMOVAL
# ═══════════════════════════════════════════════════════════
print("\n" + "=" * 60)
print("STAGE 3: OUTLIER REMOVAL")
print("=" * 60)

def remove_outliers_iqr(dataframe, column, multiplier=2.5):
    """Remove rows outside multiplier × IQR. 2.5 keeps legitimate outliers."""
    Q1  = dataframe[column].quantile(0.25)
    Q3  = dataframe[column].quantile(0.75)
    IQR = Q3 - Q1
    before = len(dataframe)
    dataframe = dataframe[
        (dataframe[column] >= Q1 - multiplier * IQR) &
        (dataframe[column] <= Q3 + multiplier * IQR)
    ]
    print(f"  {column}: removed {before - len(dataframe)} rows")
    return dataframe

df = df[df['Battery'] > 2000]
df = df[df['Mobile Height'] > 10]

for col in ['Price', 'RAM', 'Battery', 'Rear_MP']:
    df = remove_outliers_iqr(df, col, multiplier=2.5)

print(f"Rows remaining: {len(df)}")

# ═══════════════════════════════════════════════════════════
# STAGE 4 — FEATURE DROPPING
# ═══════════════════════════════════════════════════════════
print("\n" + "=" * 60)
print("STAGE 4: FEATURE DROPPING")
print("=" * 60)

# Raw text columns fully replaced by engineered features above
raw_cols = ['Model', 'Rear Camera', 'Front Camera', 'Processor', 'Colour']
df.drop(columns=[c for c in raw_cols if c in df.columns], inplace=True)

print(f"Dropped: {[c for c in raw_cols]}")
print(f"Remaining: {list(df.columns)}")

# ═══════════════════════════════════════════════════════════
# STAGE 5 — UNIVARIATE ANALYSIS
# ═══════════════════════════════════════════════════════════
print("\n" + "=" * 60)
print("STAGE 5: UNIVARIATE ANALYSIS")
print("=" * 60)

numerical_raw    = ['Price', 'RAM', 'Memory', 'Battery',
                    'Mobile Height', 'Rear_MP', 'Front_MP']
categorical_raw  = ['Brand', 'Processor_Brand', 'Processor_Tier']

# Distributions + skew / kurtosis
fig, axes = plt.subplots(3, 3, figsize=(18, 13))
axes = axes.flatten()
for i, col in enumerate(numerical_raw):
    sns.histplot(df[col], kde=True, ax=axes[i],
                 color='steelblue', edgecolor='white')
    axes[i].set_title(
        f'{col}\nSkew={df[col].skew():.2f}  Kurt={df[col].kurtosis():.2f}',
        fontsize=10
    )
for j in range(i + 1, len(axes)):
    axes[j].set_visible(False)
plt.suptitle('Univariate — Numerical Distributions',
             fontsize=14, fontweight='bold')
plt.tight_layout(); plt.show()

# Boxplots
fig, axes = plt.subplots(2, 4, figsize=(18, 8))
axes = axes.flatten()
for i, col in enumerate(numerical_raw):
    sns.boxplot(y=df[col], ax=axes[i], color='lightcoral')
    axes[i].set_title(col)
for j in range(i + 1, len(axes)):
    axes[j].set_visible(False)
plt.suptitle('Univariate — Boxplots', fontsize=14, fontweight='bold')
plt.tight_layout(); plt.show()

# Summary stats
display(Markdown("### Summary Statistics (Numerical)"))
display(df[numerical_raw]
        .agg(['mean', 'median', 'std', 'skew', 'kurtosis'])
        .T.round(2))

# Categorical value counts
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
for i, col in enumerate(categorical_raw):
    vc = df[col].value_counts()
    sns.barplot(x=vc.values, y=vc.index, ax=axes[i],
                palette='viridis', hue=vc.index, legend=False)
    axes[i].set_title(col)
plt.suptitle('Univariate — Categorical Features',
             fontsize=14, fontweight='bold')
plt.tight_layout(); plt.show()

# ═══════════════════════════════════════════════════════════
# STAGE 6 — BIVARIATE ANALYSIS
# ═══════════════════════════════════════════════════════════
print("\n" + "=" * 60)
print("STAGE 6: BIVARIATE ANALYSIS")
print("=" * 60)

# Numerical vs Price (scatter + Pearson r)
num_feats = [c for c in numerical_raw if c != 'Price']
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()
for i, col in enumerate(num_feats):
    sns.scatterplot(x=df[col], y=df['Price'], ax=axes[i],
                    alpha=0.45, color='steelblue')
    corr = df[col].corr(df['Price'])
    axes[i].set_title(f'{col} vs Price\nPearson r = {corr:.3f}')
for j in range(i + 1, len(axes)):
    axes[j].set_visible(False)
plt.suptitle('Bivariate — Numerical vs Price',
             fontsize=14, fontweight='bold')
plt.tight_layout(); plt.show()

# Categorical vs Price (boxplots ordered by median)
fig, axes = plt.subplots(1, 3, figsize=(20, 7))
for i, col in enumerate(categorical_raw):
    order = (df.groupby(col)['Price']
               .median()
               .sort_values(ascending=False)
               .index)
    sns.boxplot(x=col, y='Price', data=df, ax=axes[i],
                order=order, palette='Set2')
    axes[i].set_xticklabels(
        axes[i].get_xticklabels(), rotation=45, ha='right'
    )
    axes[i].set_title(f'{col} vs Price')
plt.suptitle('Bivariate — Categorical vs Price',
             fontsize=14, fontweight='bold')
plt.tight_layout(); plt.show()

# 5G vs Price
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
sns.boxplot(   x='Is_5G', y='Price', data=df, ax=axes[0], palette='coolwarm')
sns.violinplot(x='Is_5G', y='Price', data=df, ax=axes[1], palette='coolwarm')
sns.stripplot(
    x='Is_5G', y='Price', data=df, ax=axes[2],
    palette='coolwarm',
    jitter=True,       # spreads points horizontally so they don't stack
    alpha=0.5,         # transparency to handle overlap
    size=4             # point size
)
axes[2].set_title('Strip Plot')
for ax in axes:
    ax.set_xticklabels(['Non-5G', '5G'])
plt.suptitle('5G vs Non-5G Price', fontsize=13, fontweight='bold')
plt.tight_layout(); plt.show()


# Correlation heatmap (lower triangle only)
plt.figure(figsize=(10, 7))

num_corr = df.select_dtypes(include=['float64', 'int64']).corr()

# Drop columns AND rows that are entirely NaN (constant/low-variance columns)
num_corr = num_corr.dropna(axis=0, how='all').dropna(axis=1, how='all')

# k=1 excludes the diagonal (self-correlation is always 1, adds no info)
mask = np.triu(np.ones_like(num_corr, dtype=bool), k=1)

sns.heatmap(
    num_corr,
    annot=True,
    cmap='coolwarm',
    fmt='.2f',
    mask=mask,
    linewidths=0.5,
    vmin=-1, vmax=1,
    annot_kws={"size": 9}
)

plt.title('Correlation Heatmap (Lower Triangle)', fontsize=13, fontweight='bold')
plt.xticks(rotation=45, ha='right')   # stops x-labels from being cut off
plt.tight_layout(); plt.show()

# Pairplot (key features only for speed)
sns.pairplot(df[['Price', 'RAM', 'Memory', 'Battery', 'Rear_MP']],
             diag_kind='kde', plot_kws={'alpha': 0.45})
plt.suptitle('Pairplot — Key Features', y=1.02, fontsize=13)
plt.show()

# ═══════════════════════════════════════════════════════════
# STAGE 7 — LABEL ENCODING
# ═══════════════════════════════════════════════════════════
print("\n" + "=" * 60)
print("STAGE 7: LABEL ENCODING")
print("=" * 60)

# Applied on full dataset — LabelEncoder only maps vocabulary,
# it does not use target statistics so there is no leakage risk.
le_dict = {}
for col in categorical_raw:
    le             = LabelEncoder()
    df[col]        = le.fit_transform(df[col].astype(str))
    le_dict[col]   = le
    print(f"  {col}: {dict(zip(le.classes_, le.transform(le.classes_)))}")

# ═══════════════════════════════════════════════════════════
# STAGE 7 — LOG-TRANSFORM TARGET
# ═══════════════════════════════════════════════════════════
print(f"\nPrice skew BEFORE log: {df['Price'].skew():.3f}")
df['log_Price'] = np.log1p(df['Price'])
print(f"Price skew AFTER  log: {df['log_Price'].skew():.3f}")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.histplot(df['Price'],     kde=True, ax=axes[0], color='steelblue')
axes[0].set_title(f'Original Price  (skew={df["Price"].skew():.2f})')
sns.histplot(df['log_Price'], kde=True, ax=axes[1], color='coral')
axes[1].set_title(f'log(Price+1)  (skew={df["log_Price"].skew():.2f})')
plt.suptitle('Target — Before vs After Log Transform', fontweight='bold')
plt.tight_layout(); plt.show()

# ═══════════════════════════════════════════════════════════
# STAGE 8 — TRAIN / TEST SPLIT
# ═══════════════════════════════════════════════════════════
print("\n" + "=" * 60)
print("STAGE 8: TRAIN / TEST SPLIT")
print("=" * 60)

base_features = (
    ['Memory', 'RAM', 'Battery', 'Mobile Height',
     'Rear_MP', 'Front_MP', 'Is_5G']
    + categorical_raw           # label-encoded integers
    + ['Processor_Tier_Num']
)
if 'Has_AI_Lens' in df.columns:
    base_features.append('Has_AI_Lens')

X_base = df[base_features]
y      = df['log_Price']

X_train_base, X_test_base, y_train, y_test = train_test_split(
    X_base, y, test_size=0.2, random_state=42
)
print(f"Train: {X_train_base.shape}   Test: {X_test_base.shape}")

# ═══════════════════════════════════════════════════════════
# STAGE 9 — INTERACTION & ADVANCED FEATURE ENGINEERING
# ═══════════════════════════════════════════════════════════
print("\n" + "=" * 60)
print("STAGE 9: INTERACTION & ADVANCED FEATURE ENGINEERING")
print("=" * 60)

def add_interaction_features(X):
    X = X.copy()
    X['RAM_x_Memory']        = X['RAM'] * X['Memory']
    X['Camera_Ratio']        = X['Rear_MP'] / (X['Front_MP'] + 1)
    X['5G_x_ProcTier']       = X['Is_5G']  * X['Processor_Tier_Num']
    X['RAM_sq']              = X['RAM'] ** 2
    X['log_Memory']          = np.log1p(X['Memory'])
    X['Battery_per_Height']  = X['Battery'] / (X['Mobile Height'] + 1)
    X['RAM_x_ProcTier']      = X['RAM'] * X['Processor_Tier_Num']
    return X

# Brand frequency encoding (computed on train only)
brand_freq                       = X_train_base['Brand'].value_counts()
X_train_base                     = X_train_base.copy()
X_test_base                      = X_test_base.copy()
X_train_base['Brand_Freq']       = X_train_base['Brand'].map(brand_freq).fillna(1)
X_test_base['Brand_Freq']        = X_test_base['Brand'].map(brand_freq).fillna(1)

# Brand_Price_Tier — target encoding, train-only median
train_tmp                        = X_train_base.copy()
train_tmp['log_Price']           = y_train.values
brand_price_median               = train_tmp.groupby('Brand')['log_Price'].median()
global_median                    = y_train.median()

X_train_base['Brand_Price_Tier'] = (
    X_train_base['Brand'].map(brand_price_median).fillna(global_median)
)
X_test_base['Brand_Price_Tier']  = (
    X_test_base['Brand'].map(brand_price_median).fillna(global_median)
)

print("  Brand_Price_Tier: median log-price per brand (train-only, no leakage)")
print(f"  Top 5 brands by price tier:\n"
      f"{brand_price_median.sort_values(ascending=False).head()}")

X_train = add_interaction_features(X_train_base)
X_test  = add_interaction_features(X_test_base)

all_feat_names = list(X_train.columns)
print(f"\nTotal features: {len(all_feat_names)}")
print(f"Features: {all_feat_names}")

# ═══════════════════════════════════════════════════════════
# STAGE 10 — MUTUAL INFORMATION & VIF
# (computed on raw / unscaled train data — MI is rank-based,
#  VIF is scale-invariant after centering)
# ═══════════════════════════════════════════════════════════
print("\n" + "=" * 60)
print("STAGE 10: MUTUAL INFORMATION & VIF")
print("=" * 60)

mi_scores = mutual_info_regression(X_train, y_train, random_state=42)
mi_df     = pd.DataFrame({
    'Feature': all_feat_names,
    'MI_Score': mi_scores
}).sort_values('MI_Score', ascending=False)

display(Markdown("### Mutual Information Scores"))
display(mi_df)

plt.figure(figsize=(12, 7))
sns.barplot(x='MI_Score', y='Feature', data=mi_df,
            palette='viridis', hue='Feature', legend=False)
plt.title('Mutual Information Scores vs log(Price)',
          fontsize=13, fontweight='bold')
plt.tight_layout(); plt.show()

# VIF on standardized data (standard practice for VIF)
X_vif     = pd.DataFrame(
    RobustScaler().fit_transform(X_train),
    columns=all_feat_names
)
vif_data  = pd.DataFrame({
    'Feature': X_vif.columns,
    'VIF':     [variance_inflation_factor(X_vif.values, i)
                for i in range(len(X_vif.columns))]
})
display(Markdown("### VIF Values (>10 = multicollinearity concern)"))
display(vif_data.sort_values('VIF', ascending=False))

K_FEATURES = min(14, len(all_feat_names))

# ═══════════════════════════════════════════════════════════
# STAGE 11 — PIPELINE BUILDER
# ═══════════════════════════════════════════════════════════
# KEY DESIGN:
#   Ridge   → RobustScaler included (MUST scale for L2 penalty
#              to be fair across features; Robust handles skew)
#   RF      → No scaler (tree splits are threshold-based,
#              scale has zero effect on predictions)
#   XGBoost → No scaler
#   GBR     → No scaler
#   MI selector → No scaler needed
# ═══════════════════════════════════════════════════════════

def build_pipeline_ridge():
    """Ridge: RobustScaler → SelectKBest → Ridge."""
    return Pipeline(steps=[
        ('scaler',   RobustScaler()),                           # ✅ required
        ('selector', SelectKBest(mutual_info_regression,
                                 k=K_FEATURES)),
        ('model',    Ridge(alpha=1.0)),
    ])

def build_pipeline_tree(model):
    """Trees: SelectKBest → model. No scaler — scale-invariant."""
    return Pipeline(steps=[
        ('selector', SelectKBest(mutual_info_regression,        # ❌ no scaler
                                 k=K_FEATURES)),
        ('model',    model),
    ])

# ═══════════════════════════════════════════════════════════
# STAGE 12 — EVALUATION HELPER
# ═══════════════════════════════════════════════════════════

def evaluate(model_name, pipeline, Xtr, Xte, ytr, yte, run_cv=True):
    pipeline.fit(Xtr, ytr)
    log_pred   = pipeline.predict(Xte)
    price_pred = np.expm1(log_pred)
    price_true = np.expm1(yte)

    r2_log  = r2_score(yte,        log_pred)
    r2_orig = r2_score(price_true, price_pred)
    mae     = mean_absolute_error(price_true, price_pred)
    rmse    = np.sqrt(mean_squared_error(price_true, price_pred))

    print(f"\n{'─'*52}")
    print(f"  {model_name}")
    print(f"  Scaler:           "
          f"{'RobustScaler' if 'Ridge' in model_name else 'None (tree)'}")
    print(f"  R² (log scale):   {r2_log:.4f}")
    print(f"  R² (₹ scale):     {r2_orig:.4f}")
    print(f"  MAE:  ₹{mae:>10,.0f}")
    print(f"  RMSE: ₹{rmse:>10,.0f}")

    if run_cv:
        cv = cross_val_score(pipeline, Xtr, ytr, cv=5, scoring='r2')
        print(f"  CV R² (5-fold):   {cv.mean():.4f} ± {cv.std():.4f}")
    print(f"{'─'*52}")

    # Feature importance for tree models
    if any(m in model_name for m in
           ['Random Forest', 'XGBoost', 'Gradient Boosting']):
        sel_mask     = pipeline.named_steps['selector'].get_support()
        sel_features = [all_feat_names[i]
                        for i, m in enumerate(sel_mask) if m]
        importances  = pipeline.named_steps['model'].feature_importances_
        imp_df       = (
            pd.DataFrame({
                'Feature':    sel_features,
                'Importance': importances
            }).sort_values('Importance', ascending=False)
        )
        plt.figure(figsize=(11, 6))
        sns.barplot(x='Importance', y='Feature', data=imp_df,
                    palette='rocket', hue='Feature', legend=False)
        plt.title(f'Feature Importance — {model_name}', fontweight='bold')
        plt.tight_layout(); plt.show()

    return pipeline, r2_orig

# ═══════════════════════════════════════════════════════════
# STAGE 13 — BASE MODELS
# ═══════════════════════════════════════════════════════════
print("\n" + "=" * 60)
print("STAGE 13: BASE MODELS")
print("=" * 60)

display(Markdown("""
| Model | Scaler | Reason |
|---|---|---|
| Ridge Regression | **RobustScaler** | L2 penalty requires fair scale; Robust handles skewed features |
| Random Forest | **None** | Tree splits are scale-invariant |
| Gradient Boosting | **None** | Tree splits are scale-invariant |
| XGBoost | **None** | Tree splits are scale-invariant |
"""))

base_pipelines = {
    'Ridge Regression':  build_pipeline_ridge(),
    'Random Forest':     build_pipeline_tree(
        RandomForestRegressor(n_estimators=200, random_state=42)
    ),
    'Gradient Boosting': build_pipeline_tree(
        GradientBoostingRegressor(n_estimators=200, random_state=42)
    ),
    'XGBoost':           build_pipeline_tree(
        xgb.XGBRegressor(objective='reg:squarederror', random_state=42)
    ),
}

base_results = {}
for name, pipe in base_pipelines.items():
    _, r2 = evaluate(name, pipe, X_train, X_test, y_train, y_test)
    base_results[name] = r2

# Interactive widget
model_dd = widgets.Dropdown(
    options=list(base_pipelines.keys()), description='Model:'
)
run_btn  = widgets.Button(description='▶ Run Model', button_style='success')
out_wgt  = widgets.Output()

def on_run(b):
    with out_wgt:
        clear_output(wait=True)
        name = model_dd.value
        if name == 'Ridge Regression':
            pipe = build_pipeline_ridge()
        else:
            pipe = build_pipeline_tree(
                base_pipelines[name].named_steps['model'].__class__(
                    **{k: v for k, v in
                       base_pipelines[name]
                       .named_steps['model']
                       .get_params().items()}
                )
            )
        evaluate(name, pipe, X_train, X_test, y_train, y_test, run_cv=False)

run_btn.on_click(on_run)
display(Markdown("### Base Model Dashboard"))
display(widgets.HBox([model_dd, run_btn]))
display(out_wgt)

# ═══════════════════════════════════════════════════════════
# STAGE 14 — HYPERPARAMETER TUNING
# ═══════════════════════════════════════════════════════════
print("\n" + "=" * 60)
print("STAGE 14: HYPERPARAMETER TUNING — RandomizedSearchCV")
print("=" * 60)

# ── Tuned XGBoost (no scaler) ────────────────────────────────
xgb_param_dist = {
    'model__n_estimators':     randint(300, 800),
    'model__max_depth':        randint(3, 7),
    'model__learning_rate':    uniform(0.01, 0.1),
    'model__subsample':        uniform(0.65, 0.35),
    'model__colsample_bytree': uniform(0.65, 0.35),
    'model__min_child_weight': randint(3, 10),
    'model__gamma':            uniform(0, 0.5),
    'model__reg_alpha':        uniform(0, 2),
    'model__reg_lambda':       uniform(1, 3),
}

xgb_search = RandomizedSearchCV(
    build_pipeline_tree(
        xgb.XGBRegressor(objective='reg:squarederror',
                         random_state=42, tree_method='hist')
    ),
    xgb_param_dist,
    n_iter=40, cv=5, scoring='r2',
    n_jobs=-1, random_state=42, verbose=1
)
xgb_search.fit(X_train, y_train)
print(f"\nBest XGBoost params: {xgb_search.best_params_}")
print(f"CV R²: {xgb_search.best_score_:.4f}")
_, tuned_xgb_r2 = evaluate(
    'Tuned XGBoost', xgb_search.best_estimator_,
    X_train, X_test, y_train, y_test
)

# ── Tuned Random Forest (no scaler) ─────────────────────────
rf_param_dist = {
    'model__n_estimators':      randint(200, 600),
    'model__max_depth':         [None, 10, 20, 30, 40],
    'model__min_samples_split': randint(2, 10),
    'model__min_samples_leaf':  randint(1, 5),
    'model__max_features':      ['sqrt', 'log2', 0.5, 0.7],
}

rf_search = RandomizedSearchCV(
    build_pipeline_tree(RandomForestRegressor(random_state=42)),
    rf_param_dist,
    n_iter=40, cv=5, scoring='r2',
    n_jobs=-1, random_state=42, verbose=1
)
rf_search.fit(X_train, y_train)
print(f"\nBest RF params: {rf_search.best_params_}")
print(f"CV R²: {rf_search.best_score_:.4f}")
_, tuned_rf_r2 = evaluate(
    'Tuned Random Forest', rf_search.best_estimator_,
    X_train, X_test, y_train, y_test
)

# ── Tuned Ridge (with RobustScaler) ─────────────────────────
ridge_param_dist = {
    'model__alpha': uniform(0.01, 100),
}

ridge_search = RandomizedSearchCV(
    build_pipeline_ridge(),
    ridge_param_dist,
    n_iter=40, cv=5, scoring='r2',
    n_jobs=-1, random_state=42, verbose=1
)
ridge_search.fit(X_train, y_train)
print(f"\nBest Ridge alpha: {ridge_search.best_params_}")
print(f"CV R²: {ridge_search.best_score_:.4f}")
_, tuned_ridge_r2 = evaluate(
    'Tuned Ridge', ridge_search.best_estimator_,
    X_train, X_test, y_train, y_test
)

# ═══════════════════════════════════════════════════════════
# STAGE 15 — RESIDUAL ANALYSIS (best model)
# ═══════════════════════════════════════════════════════════
print("\n" + "=" * 60)
print("STAGE 15: RESIDUAL ANALYSIS")
print("=" * 60)

all_results = {
    **base_results,
    'Tuned XGBoost':      tuned_xgb_r2,
    'Tuned Random Forest': tuned_rf_r2,
    'Tuned Ridge':         tuned_ridge_r2,
}

best_name = max(all_results, key=all_results.get)
best_pipe = {
    'Tuned XGBoost':      xgb_search.best_estimator_,
    'Tuned Random Forest': rf_search.best_estimator_,
    'Tuned Ridge':         ridge_search.best_estimator_,
}.get(best_name, base_pipelines.get(best_name))

print(f"Best model: {best_name}  R²={all_results[best_name]:.4f}")

log_pred   = best_pipe.predict(X_test)
price_pred = np.expm1(log_pred)
price_true = np.expm1(y_test)
residuals  = price_true - price_pred

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

axes[0].scatter(price_pred, residuals, alpha=0.4, color='steelblue', s=20)
axes[0].axhline(0, color='red', linestyle='--')
axes[0].set_xlabel('Predicted Price (₹)')
axes[0].set_ylabel('Residual (₹)')
axes[0].set_title('Residuals vs Fitted')

sns.histplot(residuals, kde=True, ax=axes[1], color='coral')
axes[1].set_title(f'Residual Distribution\nSkew={residuals.skew():.2f}')

mn = min(price_true.min(), price_pred.min())
mx = max(price_true.max(), price_pred.max())
axes[2].scatter(price_true, price_pred, alpha=0.4, color='seagreen', s=20)
axes[2].plot([mn, mx], [mn, mx], 'r--', linewidth=1.5, label='Perfect fit')
axes[2].set_xlabel('Actual Price (₹)')
axes[2].set_ylabel('Predicted Price (₹)')
axes[2].set_title('Actual vs Predicted')
axes[2].legend()

plt.suptitle(f'Residual Analysis — {best_name}',
             fontweight='bold', fontsize=13)
plt.tight_layout(); plt.show()

# # ═══════════════════════════════════════════════════════════
# STAGE 16 — FINAL REPORT
# ═══════════════════════════════════════════════════════════
print("\n" + "=" * 60)
print("STAGE 16: FINAL MODEL COMPARISON REPORT")
print("=" * 60)

perf_df = pd.DataFrame({
    'Model':    list(all_results.keys()),
    'R² (₹)':  list(all_results.values()),
    'Scaler':   [
        'RobustScaler' if 'Ridge' in m else 'None'
        for m in all_results.keys()
    ],
}).sort_values('R² (₹)', ascending=False)

display(Markdown("### Model Performance"))
display(
    perf_df.style
    .background_gradient(subset=['R² (₹)'], cmap='Greens')
    .format({'R² (₹)': '{:.4f}'})
)

# Bar chart with target lines
plot_df = perf_df.sort_values('R² (₹)', ascending=True)
plt.figure(figsize=(13, 6))
bar_colors = [
    '#22c55e' if r >= 0.85
    else '#f97316' if r >= 0.80
    else '#ef4444'
    for r in plot_df['R² (₹)']
]
bars = plt.barh(plot_df['Model'], plot_df['R² (₹)'],
                color=bar_colors, edgecolor='white')
plt.axvline(0.85, color='green',  linestyle='--',
            linewidth=1.5, label='Target 0.85')
plt.axvline(0.80, color='orange', linestyle='--',
            linewidth=1.5, label='Baseline 0.80')
for bar, val in zip(bars, plot_df['R² (₹)']):
    plt.text(bar.get_width() + 0.004,
             bar.get_y() + bar.get_height() / 2,
             f'{val:.4f}', va='center',
             fontsize=11, fontweight='bold')
plt.xlim(0, 1.05)
plt.xlabel('R² Score (₹ scale)')
plt.title('Final Model Comparison — v4', fontsize=14, fontweight='bold')
plt.legend(); plt.tight_layout(); plt.show()

display(Markdown("### Mutual Information Scores"))
display(mi_df)

display(Markdown("### VIF Table"))
display(vif_data.sort_values('VIF', ascending=False))

display(Markdown("### Standardization Summary"))
display(Markdown("""
| Model | Scaler Used | Why |
|---|---|---|
| Ridge | ✅ RobustScaler | L2 penalty must penalise all features equally; Robust handles skewed distributions better than Standard |
| Random Forest | ❌ None | Decision tree splits are threshold-based — scale has zero effect |
| Gradient Boosting | ❌ None | Same as Random Forest |
| XGBoost | ❌ None | Same as Random Forest |
| SelectKBest (MI) | ❌ None | Mutual Information is rank-based, not distance-based |
| VIF computation | ✅ RobustScaler | Standard practice for numerical stability |
"""))

# ═══════════════════════════════════════════════════════════
# STAGE 17 — ORGANIZATIONAL RECOMMENDATIONS
# ═══════════════════════════════════════════════════════════

print("\n" + "=" * 60)
print("STAGE 17: ORGANIZATIONAL RECOMMENDATIONS")
print("=" * 60)

best_r2 = all_results[best_name]

print(f"\nBest Model Identified: {best_name}")
print(f"R² Score: {best_r2:.4f}")

print("\n1. MODEL DEPLOYMENT RECOMMENDATION")
print("--------------------------------------------------")

print(f"""
The {best_name} model demonstrated the best predictive performance
with an R² of {best_r2:.3f}. This model should be deployed in the
production pricing system to estimate smartphone prices based on
hardware specifications.

Recommended deployment workflow:

1. Input new smartphone specifications
2. Apply the same feature engineering steps
3. Run prediction using the trained pipeline
4. Convert log price prediction back to original price
""")

print("\n2. KEY PRICE DRIVERS IDENTIFIED")
print("--------------------------------------------------")

top_features = mi_df.head(5)

print("Top 5 features influencing smartphone price:")

print(top_features)

print("""
Key insights:

• RAM × Processor Tier interaction is one of the strongest predictors.
  This indicates that pricing is determined by the combined performance
  of RAM and processor capability rather than either factor individually.

• Brand pricing power (Brand_Price_Tier) significantly influences price.
  Some brands command a premium independent of specifications.

• Rear camera megapixels influence pricing more than front camera specs.

• 5G capability contributes to higher price segments.
""")

print("\n3. PRODUCT STRATEGY RECOMMENDATIONS")
print("--------------------------------------------------")

print("""
Based on model insights, the following product design strategies are recommended:

1. Focus on RAM + Processor upgrades
   Smartphones with higher RAM paired with premium processors
   create the strongest price increase potential.

2. Invest more in rear camera technology
   Rear camera specifications have higher influence on price perception.

3. Optimize memory configurations
   Increasing RAM produces higher price impact than increasing storage alone.

4. Prioritize 5G in mid-range and premium segments
   5G capability is associated with higher price tiers.
""")

print("\n4. BRAND PORTFOLIO INSIGHTS")
print("--------------------------------------------------")

brand_ranking = brand_price_median.sort_values(ascending=False)

print("Brands commanding highest price tiers:")
print(brand_ranking.head())

print("""
Recommendation:

• Premium brands can maintain higher margins due to brand equity.
• Budget brands should compete through better hardware specifications.
• Retailers can adjust pricing strategies based on brand positioning.
""")

print("\n5. MODEL MONITORING & MAINTENANCE")
print("--------------------------------------------------")

print("""
To maintain model performance:

1. Retrain model monthly with new smartphone releases.
2. Monitor prediction error (MAE) regularly.
3. Update processor classification rules when new chipsets launch.
4. Recalculate Brand_Price_Tier when new brands enter the market.
""")

print("\n6. DATA QUALITY IMPROVEMENTS")
print("--------------------------------------------------")

print("""
To improve prediction accuracy further:

• Standardize brand naming conventions.
• Ensure consistent camera specification formats (e.g., 50MP).
• Maintain structured processor classification data.
• Reduce missing or inconsistent specification fields.
""")

print("\n7. BUSINESS IMPACT")
print("--------------------------------------------------")

print("""
Deploying this model enables:

• Automated price estimation for new smartphones
• Data-driven pricing strategy
• Competitive market positioning
• Faster product catalog evaluation
• Reduced manual pricing effort
""")

print("\nOrganizational recommendations generated successfully.")

# ═══════════════════════════════════════════════════════════
# STAGE 18 — MODEL EXPLAINABILITY (SHAP)
# ═══════════════════════════════════════════════════════════

print("\n" + "=" * 60)
print("STAGE 18: MODEL EXPLAINABILITY WITH SHAP")
print("=" * 60)

import shap

# SHAP works best for tree models
if 'XGBoost' in best_name or 'Forest' in best_name or 'Boosting' in best_name:

    model = best_pipe.named_steps['model']
    selector = best_pipe.named_steps['selector']

    # Get selected feature names
    selected_mask = selector.get_support()
    selected_features = [all_feat_names[i] for i, m in enumerate(selected_mask) if m]

    # Transform training data
    X_train_selected = selector.transform(X_train)

    print("Generating SHAP values...")

    explainer = shap.TreeExplainer(model)
    shap_values = explainer.shap_values(X_train_selected)

    plt.figure()
    shap.summary_plot(
        shap_values,
        X_train_selected,
        feature_names=selected_features,
        show=False
    )
    plt.title("SHAP Feature Importance")
    plt.show()

else:
    print("SHAP visualization skipped — best model is linear (Ridge).")

# ═══════════════════════════════════════════════════════════
# STAGE 18 — MODEL EXPLAINABILITY (SHAP)
# ═══════════════════════════════════════════════════════════

print("\n" + "=" * 60)
print("STAGE 18: MODEL EXPLAINABILITY WITH SHAP")
print("=" * 60)

import shap

# SHAP works best for tree models
if 'XGBoost' in best_name or 'Forest' in best_name or 'Boosting' in best_name:

    model = best_pipe.named_steps['model']
    selector = best_pipe.named_steps['selector']

    # Get selected feature names
    selected_mask = selector.get_support()
    selected_features = [all_feat_names[i] for i, m in enumerate(selected_mask) if m]

    # Transform training data
    X_train_selected = selector.transform(X_train)

    print("Generating SHAP values...")

    explainer = shap.TreeExplainer(model)
    shap_values = explainer.shap_values(X_train_selected)

    plt.figure()
    shap.summary_plot(
        shap_values,
        X_train_selected,
        feature_names=selected_features,
        show=False
    )
    plt.title("SHAP Feature Importance")
    plt.show()

else:
    print("SHAP visualization skipped — best model is linear (Ridge).")

# ═══════════════════════════════════════════════════════════
# STAGE 19 — PRODUCTION PREDICTION FUNCTION
# ═══════════════════════════════════════════════════════════

print("\n" + "=" * 60)
print("STAGE 19: PRODUCTION PRICE PREDICTION FUNCTION")
print("=" * 60)

def predict_smartphone_price(phone_specs):

    df_input = pd.DataFrame([phone_specs])

    # Add missing columns automatically
    for col in X_train.columns:
        if col not in df_input.columns:
            df_input[col] = 0

    # Ensure same column order
    df_input = df_input[X_train.columns]

    log_price = best_pipe.predict(df_input)

    price = np.expm1(log_price)[0]

    return round(price,2)

    print("\nExample Prediction")

sample_phone = {

    'Memory':128,
    'RAM':8,
    'Battery':5000,
    'Mobile Height':16,
    'Rear_MP':50,
    'Front_MP':16,
    'Is_5G':1,

    'Brand':2,
    'Processor_Brand':1,
    'Processor_Tier':2,
    'Processor_Tier_Num':2,

    'Brand_Freq':20,
    'Brand_Price_Tier':9
}

predicted_price = predict_smartphone_price(sample_phone)

print(f"Predicted Smartphone Price: ₹{predicted_price:,.0f}")


# ═══════════════════════════════════════════════════════════
# STAGE 20 — MODEL DRIFT MONITORING
# ═══════════════════════════════════════════════════════════

print("\n" + "=" * 60)
print("STAGE 20: MODEL PERFORMANCE MONITORING")
print("=" * 60)

# Drift threshold
drift_threshold = mae_best * 1.5

print(f"Training MAE: ₹{mae_best:,.0f}")
print(f"Drift Alert Threshold: ₹{drift_threshold:,.0f}")

print("""
Monitoring Strategy:

• If production MAE exceeds this threshold,
  the model should be retrained.

• Monitor prediction accuracy weekly.

• Retrain model when:

    - New smartphone chipsets launch
    - Significant market price changes occur
    - MAE exceeds drift threshold
""")
```
